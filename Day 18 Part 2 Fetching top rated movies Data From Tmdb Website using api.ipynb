```python
for i in range(1, 477):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=44d9e09dc9e681c1c705e25d7cfa2f42&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'release_date', 'overview', 'popularity', 'vote_average', 'vote_count']]
    df = df.append(temp_df)
```
